2. Write a Python function called rentals_month that retrieves rental data for a given month and year (passed as parameters) from the Sakila database as a Pandas DataFrame. The function should take in three parameters:

- engine: an object representing the database connection engine to be used to establish a connection to the Sakila database.
- month: an integer representing the month for which rental data is to be retrieved.
- year: an integer representing the year for which rental data is to be retrieved.

The function should execute a SQL query to retrieve the rental data for the specified month and year from the rental table in the Sakila database, and return it as a pandas DataFrame.

```python
# Create the rentals_month() function

def rentals_month (year: int, month: int, engine) -> pd.DataFrame:
    '''
    This function retrieves rental data for a given month and year from the Sakila database as Pandas Dataframe

    Inputs:
    The inputs of this function are two integers (month and year) and the engine that makes the connection between SQL and Python

    Outputs:
    A Pandas DataFrame with rental information
    '''

    query = f"SELECT * FROM SAKILA.RENTAL WHERE EXTRACT(YEAR FROM RENTAL_DATE) = {year} AND EXTRACT(MONTH FROM RENTAL_DATE) = {month};"

    df = pd.read_sql(query, engine)
    
    return df
```

```python
rentals_month(2005, 5, engine)
```

```python
# Create dataframe

rentals_month_df = rentals_month(2005, 5, engine)
display(rentals_month_df)
```

3. Develop a Python function called rental_count_month that takes the DataFrame provided by rentals_month as input along with the month and year and returns a new DataFrame containing the number of rentals made by each customer_id during the selected month and year.

The function should also include the month and year as parameters and use them to name the new column according to the month and year, for example, if the input month is 05 and the year is 2005, the column name should be "rentals_05_2005".

```python
# Create the rental_count_month() function

def rental_count_month (df: pd.DataFrame, year: int, month: int) -> pd.DataFrame:
    '''
    This function retrieves rental data with the number of rentals per customer during a selected month and year.

    Inputs:
    The inputs of this function are two integers (month and year) and a dataframe with rental information

    Outputs:
    A Pandas DataFrame with rental information
    '''
    
    rentals_per_customer = df.groupby(['customer_id']).agg({'rental_id':'count'})
    
    new_column_name = f'rentals_{month}_{year}'
    rentals_per_customer.rename(columns={'rental_id': new_column_name}, inplace=True)
    
    return rentals_per_customer
```

```python
rental_count_month(rentals_month_df, 2005, 5)
```

4.Create a Python function called compare_rentals that takes two DataFrames as input containing the number of rentals made by each customer in different months and years. The function should return a combined DataFrame with a new 'difference' column, which is the difference between the number of rentals in the two months.

```python
# Create dataframes from rentals_month() function and display them. One should have rental data from May 2005 and the other from February 2006

rentals_may_2005 = rentals_month(2005, 5, engine)
rentals_february_2006 = rentals_month(2006, 2, engine)

display(rentals_may_2005)
display(rentals_february_2006)
```

```python
# Create new dataframes from the rental_count_month() function 

rental_count_may_2005 = rental_count_month(rentals_may_2005, 2005, 5)
rental_count_february_2006 = rental_count_month(rentals_february_2006, 2006, 2)

display(rental_count_may_2005)
display(rental_count_february_2006)
```

```python
# Create the compare_rentals() function

def compare_rentals (df1: pd.DataFrame, df2: pd.DataFrame) -> pd.DataFrame:
    '''
    This function takes two dataframes with rental data from different years and months and retrieves a combined dataframe with this data and a new column with the difference in rentals between those years and months

    Inputs:
    Two dataframes with rental information for different years and months

    Outputs:
    A consolidated Pandas dataframe with information from the inputs and a new column
    '''
    
    combined_df = pd.concat([df1, df2], axis=1)
    
    combined_df_no_nans = combined_df.fillna(0)
    
    final_df['rentals_5_2005'] = combined_df_no_nans['rentals_5_2005']
    final_df['rentals_2_2006'] = combined_df_no_nans['rentals_2_2006']
    final_df['difference'] = combined_df_no_nans['rentals_5_2005'] - combined_df_no_nans['rentals_2_2006']

    return final_df
```

```python
compare_rentals(rental_count_may_2005, rental_count_february_2006)
```
